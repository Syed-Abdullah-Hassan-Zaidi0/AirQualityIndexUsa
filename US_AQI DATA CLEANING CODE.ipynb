```python
import pandas as pd
import time

# Specify the file path
file_path = r'C:\Users\Abdullah\Downloads\Compressed\DATA\US_AQI.csv'

# Measure the time taken to load the dataset
start_time = time.time()
df = pd.read_csv(file_path)
end_time = time.time()

# Calculate and print the load time
load_time = end_time - start_time
print(f"Time taken to load the data: {load_time:.2f} seconds")

# Check for missing values
missing_values = df.isnull().sum()

# Check for duplicate rows
duplicate_rows = df.duplicated().sum()

# Get summary statistics for numeric columns only
summary_stats = df.describe()

# Check data types
data_types = df.dtypes

# Select only numeric columns for outlier detection
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Check for outliers (using IQR method)
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).sum()

# Print the results
print("Missing Values:\n", missing_values)
print("------------------------------- \n")
print("\nDuplicate Rows:\n", duplicate_rows)
print("------------------------------- \n")
print("\nSummary Statistics:\n", summary_stats)
print("------------------------------- \n")
print("\nData Types:\n", data_types)
print("------------------------------- \n")
print("\nOutliers Detected:\n", outliers)
print("------------------------------- \n")
```
